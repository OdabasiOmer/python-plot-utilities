### This example shows how to plot lines with upper/lower error bounds

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import plot_utils as pu
```

```python
%matplotlib inline
```

### Generate data points

```python
x = np.arange(0,10,0.01)
y0 = 10 * np.sin(x)  # "clean" data
noise = np.random.normal(0,1,len(x))
error = np.random.normal(6,0.5,len(x))
y1 = y0 + noise   # data with "noises"
y1_upper = y1 + error  # "measurement errors", upper bound
y1_lower = y1 - error  # "measurement errors", lower bound
```

### Plot data with error bounds

Plot with default arguments:

```python
pu.plot_with_error_bounds(x,y1,y1_upper,y1_lower);
```

The user can also add additional lines, adjust legends, etc.

```python
fig,ax = pu.plot_with_error_bounds(x,y1,y1_upper,y1_lower,line_label='Experiment',shade_label='Error bounds')

ax.plot(x,y0,'r',label='Theory')  # add an additional line on top
ax.legend(loc=3)
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Signal strength');
```
